```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from collections import OrderedDict
import re
import nltk
from nltk.tokenize import sent_tokenize
```

```python
# Load pre-trained GPT model and tokenizer
model_name ="gpt2"  # or any other GPT variant
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

```python
def calculate_perplexity(text):
    # Tokenize the text
    input_ids = tokenizer.encode(text, return_tensors="pt")

    # Get model perplexity
    perplexity = model.forward(input_ids, labels=input_ids)["loss"].exp().item()

    return perplexity
```

```python
def getResults( perplexity):
        threshold1=60
        threshold2=80
        if perplexity < threshold1:
            label = 0
            return "The Text is generated by AI.", label
        elif  perplexity < threshold2:
            label = 0
            return "The Text is most probably contain parts which are generated by AI. (require more text for better Judgement)", label
        else:
            label = 1
            return "The Text is written by Human.", label
```

```python
def predict(sentence):
        results = OrderedDict()

        lines = sent_tokenize(sentence)
        Perplexity_per_line = []
        for line in lines:
            ppl =calculate_perplexity(line)
            print(ppl,line)
            Perplexity_per_line.append(ppl)
        print(f"Average Perplexity of whole paragraph  {sum(Perplexity_per_line)/len(Perplexity_per_line)}")
        results["Average Perplexity of whole paragraph"] = sum(Perplexity_per_line)/len(Perplexity_per_line)


        out, label = getResults(results["Average Perplexity of whole paragraph"])
        results["label"] = label

        return out
```
